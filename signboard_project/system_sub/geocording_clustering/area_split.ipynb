```python
import glob
import numpy as np
import pandas as pd
import folium
from sklearn.cluster import KMeans
```

```python
#dataset make
file_dir = glob.glob("../../signboard_classifier/osaka_img/osaka_all_img/osaka*/*_geo.txt")

f = open('dataset.csv', 'w',encoding='UTF-8')
f.write("店舗,緯度,経度\n")

for file in file_dir:
    f2 = open(file,"r",encoding='UTF-8')
    idokedo = f2.readlines()    
    f.write(file.rsplit("\\",2)[1] + "," + idokedo[0].replace("\n","") + "," + idokedo[1] + "\n")
    
f.close()
```

```python
df = pd.read_csv('dataset.csv')

#clustering
cust_array = np.array([df['緯度'].tolist(),
                       df['経度'].tolist()])
cust_array = cust_array.T
kclusters = 18 
pred = KMeans(n_clusters=kclusters).fit_predict(cust_array)
print(pred)

df['cluster_id'] = pred
```

```python
#mapping
colors_array = ['red','brown','green',"blue","yellow","gray","black","purple","gold","orange","white","pink","silver"
               ,"lime","teal","aqua","fuchsia","olive"] 

latitude = 34.702495
longitude = 135.552673
map_3 = folium.Map(location=[latitude,longitude], zoom_start=10)

for lat, lon, cluster in zip(df['緯度'], df['経度'], df['cluster_id']):
    folium.CircleMarker(
        [lat, lon],
        radius=3,
        popup=cluster,
        color=colors_array[cluster],
        fill=True,
        fill_color=colors_array[cluster],
        fill_opacity=0.7).add_to(map_3)

map_3
```

```python
counter = [0]*kclusters
ct = 0
for i in counter:
    counter[ct] = np.sum(pred==ct)
    ct += 1
print(counter)
```

```python
df.to_csv('dataset_clustering_18.csv') 
```
